# 01 — Data Exploration

Explore raw video counts, class balance, and visual sample quality before training.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

splits_dir = Path('../data/splits')
dfs = {s: pd.read_csv(splits_dir / f'{s}.csv') for s in ['train', 'val', 'test'] if (splits_dir / f'{s}.csv').exists()}

for name, df in dfs.items():
    print(f'{name}: {len(df)} samples — real: {(df.label==0).sum()}, fake: {(df.label==1).sum()}')
```

```python
# Label distribution bar chart
all_df = pd.concat(dfs.values())
all_df.groupby(['split', 'label']).size().unstack().plot(kind='bar', figsize=(8, 4), title='Frame counts per split')
plt.tight_layout()
plt.show()
```
